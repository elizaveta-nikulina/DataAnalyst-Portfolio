```python
df_investments['outlier'] = False #—Å—Ç–æ–ª–±–µ—Ü –¥–ª—è –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—Å–≤ 
market_unique = df_investments['market'].unique()

for x in market_unique: #–¥–ª—è –∫–∞–∂–¥–æ–π —É–Ω–∏–∫–∞–ª—å–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ –º–∞—Ä–∫–µ—Ç
    # –≤—ã–±–æ—Ä–∫–∞ —Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
    data = df_investments[df_investments['market'] == x]['funding_total_usd']
    
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    upperr = q3 + 1.5*iqr
    
    #—Å–æ–∑–¥–∞–µ–º —É—Å–ª–æ–≤–∏—è –¥–ª—è –≤—ã–±—Ä–æ—Å–æ–≤ –∏ –ø—Ä–∏–º–µ–Ω—è–µ–º —ç—Ç–æ –≤—Å–µ –∫ –¥–∞—Ç–∞—Å–µ—Ç—É
    outliers = (df_investments['market'] == x) & (df_investments['funding_total_usd'] > upperr)
    df_investments.loc[outliers, 'outlier'] = True
```
